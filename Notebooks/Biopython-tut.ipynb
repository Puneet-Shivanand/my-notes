Types of Sequences 

```python
from Bio.Seq import Seq
from Bio.Alphabet import IUPAC
prot_A = Seq("ATAGGTAAG", IUPAC.protein)
print prot_A, prot_A.alphabet
prot_B = Seq("ATAGGTAAG", IUPAC.ambiguous_dna)
print prot_B, prot_B.alphabet
prot_C = Seq("ATAGGTAAG", IUPAC.ambiguous_rna)
print prot_C, prot_C.alphabet
```

```python
print prot_A.count('AT')
print prot_A.count('TA')
```
